```python
import pandas as pd
df = pd.read_csv('fulldata_brazil.csv')
df
```

```python
review = df.review_comment_message.dropna().drop_duplicates().to_list()
review = [cmt.replace('\r',' ').replace('\n',' ') for cmt in review]
review
```

```python
len(review)
```

```python
!pip install googletrans==4.0.0rc1
from googletrans import Translator, constants
from pprint import pprint
# init the Google API translator
translator = Translator()
```

```python
# translate more than a phrase
# trans_review = [translator.translate(cmt, src="pt").text for cmt in review]
# print(trans_review)
trans_rv = list()
for translation in review:
    result = translator.translate(translation, src="pt")
    print(result.text)
    trans_rv.append(result.text)
```

```python
len(trans_rv)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#convert list to string and generate
unique_string=(" ").join(trans_rv)
wordcloud = WordCloud(width = 1000, height = 500).generate(unique_string)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("all_review"+".png", bbox_inches='tight')
plt.show()
plt.close()
```
