```python
# create new feature: no. waiting days
df.order_delivered_customer_date = pd.to_datetime(df['order_delivered_customer_date'])
df.order_purchase_timestamp = pd.to_datetime(df['order_purchase_timestamp'])
df['waiting_days'] = (df.order_delivered_customer_date - df.order_purchase_timestamp)
```

```python
df['waiting_days'] = df.waiting_days.dt.days
```

```python
df.waiting_days.dt.days.hist(bins=50)
```

```python
# Visualize kernel density estimate
good = df.review_score >=4
bad = df.review_score <=2
def kdeplot(feature):
    ax = sns.kdeplot(df[feature][good], color = "Blue", shade = True);
    ax = sns.kdeplot(df[feature][bad], color = "Red", shade = True);

    ax.legend(["good", "bad"], loc = "upper right");
    ax.set_ylabel("Density");
    ax.set_xlabel("{}".format(feature));
    ax.set_title("Distribution of {} by review score".format(feature));
    plt.show();

train_columns = ['price','freight_value','payment_value','waiting_days']
for feature in train_columns:
    kdeplot(feature)
```

KDE plot is a probability density function that generates the data by binning and counting observations. But, rather than using a discrete bin KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate. KDE can produce a plot that is less cluttered and more interpretable, especially when drawing multiple distributions. However, sometimes the KDE plot has the potential to introduce distortions if the underlying distribution is bounded or not smooth.
